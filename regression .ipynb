```python
data.describe()
```

```python
sns.boxplot(x='sex', y='age', data=data)
sns.despine()
```

```python
sns.boxplot(x='sex', y='bmi', data=data)
sns.despine()
```

```python
sns.boxplot(x="sex",y="expenses",data=data)
sns.despine
```

```python
plt.hist(data['age'])
plt.title('Histogram of age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
data.groupby('region')['expenses'].mean().plot(kind='bar')
plt.title('Average expenses by region')
plt.xlabel('Region')
plt.ylabel('Average expenses')
plt.show()
```

```python
sns.countplot(x='sex', hue='smoker', data=data)
sns.despine()
```

```python
sns.countplot(x=data.sex)
```
