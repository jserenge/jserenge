```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data_wed=pd.read_excel('social.xlsx',sheet_name='Edges')
```

```python
data_wed.head()
```

```python
import networkx as nx
```

```python
G = nx.from_pandas_edgelist(data_wed,source='Source', target='Target',create_using=nx.DiGraph())
                            
```

```python
print(nx.info(G))
```

```python
plt.figure(num=None, figsize=(20, 20), dpi=80)
plt.axis('off')
fig = plt.figure(1)
nx.draw_networkx(G)
```

```python
# using the betweeness centrality to give each node a different size 
```

```python
pos = nx.spring_layout(G)
betCent = nx.betweenness_centrality(G, normalized=True, endpoints=True)
node_color = [20000.0 * G.degree(v) for v in G]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
```
